# Compute Objective functions

```python
# Contains first solution of every run times of every datasets
df_first = pd.DataFrame(columns=['Dataset No.','Run No.','Power weight','Num. weight','Dist. weight','Power','Num. Active Sensors','Avg. Dist. to Sink node'])
# Contains last solution of every run times of every datasets
df_last = pd.DataFrame(columns=['Dataset No.','Run No.','Power weight','Num. weight','Dist. weight','Power','Num. Active Sensors','Avg. Dist. to Sink node'])
```

```python
dataset = 0
with open(f'Datasets/uniform/50x1000unit/300sensors/sensors_positions_{dataset}.pickle','rb') as file:
    sensors_positions = pickle.load(file)
with open('Datasets/sink_nodes_positions.pickle','rb') as file:
    sink_nodes_positions = pickle.load(file) 
with open(f'MOEAD_Results/uniform/50x1000unit/300sensors/dataset_{dataset}/lambdas_{dataset}.pickle','rb') as file:
    lambdas = pickle.load(file)

distances = np.zeros(shape=(300, 300))
for i in range(300):
    for j in range(300):
        d = MOEAD.Individual.euclid_distance(sensors_positions[i], sensors_positions[j])
        distances[i,j] = distances[j,i] = d

for runtime in range(1):
    with open(f'MOEAD_Results/uniform/50x1000unit/300sensors/dataset_{dataset}/last_solutions_{runtime}.pickle','rb') as file:
        last_sol = pickle.load(file)
    with open(f'MOEAD_Results/uniform/50x1000unit/300sensors/dataset_{dataset}/first_solutions_{runtime}.pickle','rb') as file:
        first_sol = pickle.load(file)

    for i in range(len(first_sol)):
        active_indx = [[],[]]
        for s in range(len(sensors_positions)):            
            if(first_sol[i][s][0]==1):
                active_indx[0].append(s)
            if(last_sol[i][s][0]==1):
                active_indx[1].append(s)
        first_indi = MOEAD.Individual(lambdas[i], 300, 1, sensors_positions, sink_nodes_positions, [0,0,0], [1e9,1e9,1e9], distances, solution=first_sol[i], active_indx=active_indx[0])
        first_indi.compute_fitness(first_sol[i], [0,0,0], [1e9,1e9,1e9])

        last_indi = MOEAD.Individual(lambdas[i], 300, 1, sensors_positions, sink_nodes_positions, [0,0,0], [1e9,1e9,1e9], distances, solution=last_sol[i], active_indx=active_indx[1])
        last_indi.compute_fitness(last_sol[i], [0,0,0], [1e9,1e9,1e9])

        df_first.loc[len(df_first)] = [dataset, runtime, lambdas[i][0],lambdas[i][1],lambdas[i][2], first_indi.f[0], first_indi.f[1], first_indi.f[2]]
        df_last.loc[len(df_last)] = [dataset, runtime, lambdas[i][0],lambdas[i][1],lambdas[i][2], last_indi.f[0], last_indi.f[1], last_indi.f[2]]
```
